```python
import datetime as dt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


import cufflinks as cf

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
cf.go_offline()
init_notebook_mode(connected=True)
# import warnings
# warnings.filterwarnings('ignore')

%matplotlib inline

#stock feed data is mostly deprecated in 2017 by yahoo, google and many others
#pandas-datareader is broken
#substitute historical pickle file for feed
```

```python
bank_stocks =pd.read_pickle('all_banks')
#pickle file is python specific
```

```python
#2006 data to 2015 data
bank_stocks.head()
```

```python
#cross section slice for multiindex dataframes
#get max close price from all bank stocks
#bank_stocks['BAC']['Close'].max()
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
# create empty dataframe to store the returns of bank stocks: price at date / prev date -1
returns = pd.DataFrame()
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
for tick in tickers:
    returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

```python
#first row of data has nothign to compare it to so it it NaN; start at index 1
sns.pairplot(returns[1:])
```

```python
#november 2008 citigroup stock crash due to insolvency that is why returns are flat
```

```python
#who had the best and worst single day return
#returns['BAC Return'] ; do this for every column or idx min to get the worst days
returns.idxmin()

#20th of january is obamas inaguration day
```

```python
returns.idxmax()
#citigroups best and worst return days close to eachother effective may 6th 2011; reverse stock split
```

# rolling 30 day average against close of Bank of America

```python
plt.figure(figsize=(12,6))
bank_stocks['BAC']['Close'].ix['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 day moving average')
bank_stocks['BAC']['Close'].ix['2008-01-01':'2009-01-01'].plot(label='BAC Close')
plt.legend()
```
