## We will try a Random Sample Concensus from scikit learn.

# RANDSAC algorithm
- fits a model from a random subset of inliers, it samples and then tries to exclude points from sample.
- we will attempt to create the minimum subset of sample points and then test to see if our training data is valid.
which will be from base_estimator.fit()
- (base_estimator.predict(X)-y) will be used to determine subsets of inliers and all data outside fo that subset will be considered outliers.
- we then add outliers to the base_estimator and the the model which contains the largest number of points while still fitting to the data will be considered our maximun best fit. 

```python
X3 = data['RM'].values.reshape(-1,1)
y3 = data['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X3,y3)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_X.reshape(-1, 1))
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,10));
plt.scatter(X[inlier_mask], y[inlier_mask], 
            c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
            c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('average number of rooms per dwelling')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc='upper left');
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

## lets look at the LSTAT thruough this model 

```python
X = data['LSTAT'].values.reshape(-1,1)
y = data['MEDV'].values
ransac.fit(X, y)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
line_X = np.arange(0, 40, 1)
line_y_ransac = ransac.predict(line_X.reshape(-1, 1))
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,10));
plt.scatter(X[inlier_mask], y[inlier_mask], 
            c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
            c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('% lower status of the population')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc='upper right')
```
