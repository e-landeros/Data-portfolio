```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
# look at data summary; almost 100k entries; timestamp is object should be datetime
df.info()
```

```python
df.head()
```

```python
df['zip'].value_counts().head(10)
```

```python
df['twp'].value_counts().head(5)
```

```python
df['title'].nunique()
```

```python
#split the title to a new column reason to hold the codes in string value.
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
#verify our new column in Reason Column
df.head()
```

```python
# what is the most common reasons for 911 calls
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=df,palette='viridis')
```

```python
#time stamps are stored as string convert them to datetime objects
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
#split up the info in time stamp for better analysis
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
df.head()
```

```python
#visual representation of call type and day of the week
sns.countplot(x='Day of Week',data=df,hue='Reason',palette='viridis')
#change size of plot
sns.set_context('talk')
#move legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
#new column with stripped date
df['Date']=df['timeStamp'].apply(lambda t: t.date())
```

```python
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
#group data by day of the month and occurences
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```
