```python
#convert date colum to datetime object
df['Date'] = pd.to_datetime(df['Date'])
#format arg takes string of date 
```

```python
df.info()
```

```python
df.set_index('Date',inplace=True)
```

```python
df.head()
```

# resample method 

```python
#resample uses alias so u can group datetiem objects like groupby method
#B: business day frequency
#W: weekly
#A: year end 
df.resample(rule='A').mean()
```

```python
df.resample(rule='Q').mean()
```

```python
df.resample(rule='A').max()
```

```python
def first_day (entry):
    return entry[0]
```

```python
df.resample('A').apply(first_day)
#returns first entry for particular time period
```

# visualization

```python
df['Close'].resample('A').mean().plot(kind='bar')
```

```python
df['Close'].resample('M').mean().plot(kind='bar',figsize=(16,6))
```
