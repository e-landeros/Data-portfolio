```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# rolling mean or moving average


```python
df = pd.read_csv('time_data/walmart_stock.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
#noisy daily data open prices
```

```python
#window is size of obs used for calculating
df.rolling(window=7).mean().head(14)
#first 6 values cant be calculated
#7th is mean of the first 7 
```

```python
df['Open'].plot()
df.rolling(window=30).mean()['Close'].plot(figsize=(16,8))
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA','Close']].plot(figsize=(16,8))
```

```python
df['Close'].expanding().mean().plot(figsize=(16,8))
#this plot represents value of everything before it averaged out
#avg of all values before it
```

# Bollinger bands


```python
# Close 20 MA
df['Close: 20 DM'] = df['Close'].rolling(20).mean()
# Upper = 20MA + 2*std(20)
df['Upper'] = df['Close: 20 DM'] + 2 *(df['Close'].rolling(20).std())
# Lower = 20MA - 2* std(20)
df['Lower'] = df['Close: 20 DM'] - 2 *(df['Close'].rolling(20).std())
# Close
df[['Close','Close: 20 DM', 'Upper', 'Lower']].plot(figsize=(16,6))
```
